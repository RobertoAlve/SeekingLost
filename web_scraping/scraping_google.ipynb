```python
import requests
import os
import re

# Função para limpar o nome do arquivo
def clean_filename(url):
    # Extrair o nome do arquivo da URL
    filename = url.split('/')[-1]
    # Remover caracteres inválidos
    filename = re.sub(r'[<>:"/\\|?*]', '', filename)
    return filename

# Função para baixar imagens
def download_image(url, folder, i):
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Tentar baixar a imagem
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP 4xx/5xx

        file_name = os.path.join(folder, f'Image_{i}.jpg')
        
        with open(file_name, 'wb') as file:
            file.write(response.content)
        
        print(f'Imagem salva: {file_name}')
    except Exception as e:
        print(f'Erro ao baixar {url}: {e}')
```

```python
def download_images(query, limit, output_dir, actor):
    my_downloader = simp.Downloader()
    my_downloader.search_urls(query.replace(" ", "+" ),limit=limit, verbose=True)
    urls = my_downloader.get_urls()
    print(urls)
    # Baixar imagens
    i = 1
    for url in urls:
        path = os.path.join(output_dir, actor)
        download_image(url, path, i)
        i = i + 1
```

```python
def main(list_actors, num_imgs):
    if not os.path.exists('imagens'):
        os.makedirs('imagens')

    for actor in list_actors:
        download_images(actor, num_imgs, './imagens', actor)
```

```python
file_name = "list_actors.json"
actors = []

# Ler o conteúdo do arquivo de texto de volta para um array de strings
with open(file_name, "r") as file:
    actors = file.read().splitlines()

if __name__ == '__main__':
    main(actors, 400)
```

```python
import os

def rename_images_in_range(directory, start_range, end_range, start_value):
    """
    Renomeia as imagens dentro de um diretório que contenham um número no nome em um determinado intervalo.

    Args:
        directory (str): Caminho para o diretório onde estão as imagens.
        start_range (int): Início do intervalo de números.
        end_range (int): Fim do intervalo de números.
        start_value (int): Valor numérico inicial para renomear as imagens.
    """
    # Verifica se o diretório existe
    if not os.path.exists(directory):
        print("Diretório não encontrado.")
        return

    current_value = start_value  # Inicializa o valor para renomeação

    # Verifica se o arquivo é uma imagem e se o nome contém um número no intervalo desejado
    for num in range(start_range, end_range):
        # Percorre todos os arquivos no diretório
        for filename in os.listdir(directory):
            if f'Image_{num}.' in filename:
                print("File: ", filename)
                print("Numero: ", num)
                # Extrai a extensão da imagem
                file_extension = os.path.splitext(filename)[1]
                # Define o novo nome com o valor numérico atual
                new_name = f"Image_{current_value}{file_extension}"
                # Define o caminho completo antigo e novo
                old_path = os.path.join(directory, filename)
                new_path = os.path.join(directory, new_name)

                # Renomeia a imagem
                os.rename(old_path, new_path)
                print(f"Renomeado {filename} para {new_name}")

                # Incrementa o valor numérico para o próximo nome
                current_value += 1

# Exemplo de uso
directory_path = "./images/denzel_washington"
rename_images_in_range(directory_path, start_range=35, end_range=72, start_value=81)
```
